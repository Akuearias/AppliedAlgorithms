## **E. Train a Linear Regression model using the training data with four-fold cross-validation and with appropriate evaluation metric. Do this with a closed-form solution (using the Normal Equation or SVD) and with SGD. For SGD Perform Ridge, Lasso and Elastic Net regularization – try three values of penalty term and describe its impact. Explore the impact of other hyperparameters, like batch size and learning rate (no need for grid search). Describe your findings. For SGD, display the training and validation loss as a function of training iteration. [10 points]**

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import math

necessary_columns = X_cols
selected_columns = X_cols

dataset = dataset.dropna(subset=necessary_columns)
```

```python
# Without 4-fold Cross Validation

X = dataset[selected_columns].to_numpy()
X = np.c_[np.ones(X.shape[0]), X]
y = dataset[y_cols].to_numpy().reshape(-1, 1)

theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

print(theta)
```

```python
# With 4-fold Cross Validation

k_fold = KFold(n_splits=4, shuffle=True, random_state=2)
train_mse_list = []
validation_mse_list = []

for train, test in k_fold.split(X):
  Xtrain, Xtest = X[train], X[test]
  ytrain, ytest = y[train], y[test]
  theta = np.linalg.inv(Xtrain.T.dot(Xtrain)).dot(Xtrain.T).dot(ytrain)

  ypred_train = Xtrain.dot(theta)
  ypred_test = Xtest.dot(theta)

  train_MSE = mean_squared_error(ytrain, ypred_train)
  train_mse_list.append(train_MSE)
  validation_MSE = mean_squared_error(ytest, ypred_test)
  validation_mse_list.append(validation_MSE)

average_train_mse = np.mean(train_mse_list)
average_validation_mse = np.mean(validation_mse_list)

print(f"Average training MSE: {average_train_mse}")
print(f"Average validation MSE: {average_validation_mse}")
print(f"Theta: {theta}")
```

```python
# Extract those columns as independent variable: Country name dummy,
# Log GDP per capita, Social support, Freedom to make life choices, Generosity,
# Perceptions of corruption, Positive affect, Negative affect

scaler_X = StandardScaler() # Standarization of X

X = scaler_X.fit_transform(X)

scaler_y = StandardScaler() # Standarization of y

y = scaler_y.fit_transform(y).flatten()

k_fold = KFold(n_splits=4, shuffle=True, random_state=2) # K-fold where K is 4
mse_list = []

regs = ['l2', 'l1', 'elasticnet'] # Regulation criteria
lambdas = [(x+1) * 0.001 for x in range(4, 7)] # Penalty values
learning_rates = [10**(-x) for x in range(3, 6)] # Learning rates
batch_sizes = [5, 10, 20] # Batch sizes

results = {}
best_model = None # Used for storing the best model
prev_val_MSE = math.inf # Used for storing smallest validation error

for reg in regs:
  for alpha in lambdas:
      for lr in learning_rates:
          train_mse_list = []
          val_mse_list = []

          n_epoch = 25
          for batch in batch_sizes:
            for train, test in k_fold.split(X):
              X_train, X_test = X[train], X[test]
              y_train, y_test = y[train], y[test]

              model = SGDRegressor(penalty=reg,
                              alpha=alpha,
                              learning_rate='constant',
                              eta0=lr,
                              shuffle=True,
                              max_iter=1000,
                              tol=1e-4,
                              random_state=40)

              num_samples = X_train.shape[0]
              indices = np.arange(num_samples)

              for epoch in range(n_epoch):
                np.random.shuffle(indices)
                for i in range(0, X_train.shape[0], batch):
                  batches = indices[i:i+batch]
                  Xbatched = X_train[batches]
                  ybatched = y_train[batches]
                  model.partial_fit(Xbatched, ybatched)

                  ypred_train = model.predict(X_train)
                  ypred_test = model.predict(X_test)

              train_MSE = mean_squared_error(y_train, ypred_train)
              train_mse_list.append(train_MSE)
              val_MSE = mean_squared_error(y_test, ypred_test)

              # If the validation error is smaller than the previous one, save the model and validation error.
              if val_MSE < prev_val_MSE:
                prev_val_MSE = val_MSE
                best_model = model
              val_mse_list.append(val_MSE)

            average_train_mse = np.mean(train_mse_list)
            average_val_mse = np.mean(val_mse_list)
            results[(reg, alpha, lr, batch)] = (average_train_mse, average_val_mse)
            print(f"Current iteration: SGD = {reg}, alpha = {alpha}, learning rate = {lr}, batch size = {batch}. The training MSE is {average_train_mse}, validation MSE is {average_val_mse}.")


best = min(results.items(), key=lambda x: x[1][1])
print("The best case: Regularization method: {}; alpha: {}; learning rate: {}; batch: {}; minimum validation MSE: {}".format(best[0][0], best[0][1], best[0][2], best[0][3], best[1][1]))


```

```python
best_model.coef_[0] = best_model.intercept_[0]

print(best_model.intercept_)
print(best_model.coef_)

# Record the vector of linear regression after SGD training
LR_coef = best_model.coef_
```

Obviously, what matters the most for linear regression training via SGD is the learning rate, while all other factors matter mildly.

## **F. Repeat everything from part E with polynomial regression and using SGD. Using validation loss, explore if your model overfits/underfits the data and report your findings. [10 points]**

```python
from sklearn.preprocessing import PolynomialFeatures

# Degrees of polynomial regression
degrees = [2, 3, 4]

# 4-fold cross validation
k_fold = KFold(n_splits=4, shuffle=True, random_state=2)

# Record the validation MSE
prev_val_MSE = math.inf

x_test_pol = None
best_model = None

for degree in degrees:
  pol = PolynomialFeatures(degree=degree)
  X_poly = pol.fit_transform(X)
  Xtest_poly = pol.transform(Xtest)

  assert X_poly.shape[1] == Xtest_poly.shape[1]

  pol_losses = {}

  for reg in regs:
    for alpha in lambdas:
      for lr in learning_rates:
          train_mse_list = []
          val_mse_list = []

          n_epoch = 25
          for batch in batch_sizes:
            for train, test in k_fold.split(X_poly):
              X_train_poly, X_test_poly = X_poly[train], X_poly[test]
              y_train, y_test = y[train], y[test]

              model = SGDRegressor(penalty=reg,
                              alpha=alpha,
                              learning_rate='constant',
                              eta0=lr,
                              shuffle=True,
                              max_iter=1000,
                              tol=1e-4,
                              random_state=40)

              num_samples = X_train_poly.shape[0]
              indices = np.arange(num_samples)

              for epoch in range(n_epoch):
                np.random.shuffle(indices)
                for i in range(0, X_train_poly.shape[0], batch):
                  batches = indices[i:i+batch]
                  Xbatched = X_train_poly[batches]
                  ybatched = y_train[batches]
                  model.partial_fit(Xbatched, ybatched)

                  ypred_train = model.predict(X_train_poly)
                  ypred_test = model.predict(X_test_poly)

            train_MSE = mean_squared_error(y_train, ypred_train)
            train_mse_list.append(train_MSE)
            val_MSE = mean_squared_error(y_test, ypred_test)

              # If the validation error is smaller than the previous one, save the model and validation error.
            if val_MSE < prev_val_MSE:
              prev_val_MSE = val_MSE
              best_model = model
              x_test_pol = Xtest_poly
            val_mse_list.append(val_MSE)

            average_train_mse = np.mean(train_mse_list)
            average_val_mse = np.mean(val_mse_list)
            pol_losses[(degree, reg, alpha, lr, batch)] = (average_train_mse, average_val_mse)
            print(f"Current iteration: Degree={degree}, SGD = {reg}, alpha = {alpha}, learning rate = {lr}, batch size = {batch}. The training MSE is {average_train_mse}, validation MSE is {average_val_mse}.")

best = min(pol_losses.items(), key=lambda x: x[1][1])
print("The best case: Degree: {}; Regularization method: {}; alpha: {}; learning rate: {}; batch: {}; minimum validation MSE: {}".format(best[0][0], best[0][1], best[0][2], best[0][3], best[0][4], best[1][1]))
```

We can find that the model begins to underfit in some cases of degree increase (for example from 2 to 3, using l2 regularization, alpha=0.005, learning rate = 0.001, batch size = 5), and even overfits when the degree changes. However, as the total number of degrees within our code is small, we are unable to summarize the regulations of when underfitting and overfitting will occur.
